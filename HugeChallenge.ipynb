```python
import pandas as pd
import numpy as np

path =r'D:/Pruebas Tecnicas/input.txt'
path2="D:/Pruebas Tecnicas/output.txt"
fileOutput = open(path2,'w')

data=pd.read_csv(path,header=None)

data
```

```python
array=data.values
Strarray= []
for row in array :
    stringRow=str(row)
    stringRow=stringRow.replace("[","")
    stringRow=stringRow.replace("]","")
    stringRow=stringRow.replace("'","")
    Strarray.append(stringRow)
    print(stringRow)
    
  
```

```python
def createHorizontalEdge(w):
    fileOutput.write('-' * w)
    fileOutput.write("\n")
    print('-' * w)
```

```python
def createVerticalEdges(w,h):
    
    #for n in range(h):
        #print('|'+" "*spaces+'|')
    arr=[[' ' for j in range(w+2)] for i in range(h)]
    for j in range(h):
        arr[j][0]='|'
        arr[j][w+1]='|'   
    printMatrix(arr)
    return arr
    
    
```

```python
def createCanvas ( row ):
        print("width"+ row[1])
        print("height"+row[2])
        width=int(row[1])
        height=int(row[2])
        createHorizontalEdge(width)
        arr=createVerticalEdges(width,height)
        createHorizontalEdge(width)
        print(" ")
        return width,height,arr
      
    
```

```python
def validations(x1,x2,y1,y2,width,height ):
     if(x2>width or y2>height  or x1>width or y1>height ):
        fileOutput.write("the line written is outside the dimmension please write dimensions inside the canvas")
        print ("the line written is outside the dimmension please write dimensions inside the canvas")
        fileOutput.close()
     if(x1>x2 ) :
        fileOutput.write("x1: "+str(x1)+" must be  lower than  x2: "+str(x2))
        print("x1: "+str(x1)+" must be  lower than  x2: "+str(x2))
        fileOutput.close()
     if(y1>y2) :
        fileOutput.write("x1: "+str(y1)+" must be  lower than  x2: "+str(y2))
        print("y1: "+str(y1)+" must be  lower than  y2: "+str(y2))
        fileOutput.close()
```

```python
def createLineInside(row, width, height,arr):
    x1=int(row[1])
    y1=int(row[2])
    x2=int(row[3])
    y2=int(row[4])
    if(x1 !=x2 and y1!=y2):
        print("it is only  allowed  vertical or horizontal lines please put a valid values" )
        fileOutput.write("it is only  allowed  vertical or horizontal lines please put a valid values")
        fileOutput.close()
    validations(x1,x2,y1,y2,width,height )
    y1=y1-1
    arr1=arr
    for x in range(y1,y2):
        for y  in range(x1,x2+1):
            print("x"+str(x+1),"y"+str(y))
            arr1[x][y]='X'
    printMatrix(arr)
    return arr
```

```python
def printMatrix(arr) :
    for row in arr:
        for elem in row:
            fileOutput.write(elem) 
            print(elem , end=" ")
        fileOutput.write("\n") 
        print(" ")
    
```

```python
def createLine (row, width, height ,arr) :
    createHorizontalEdge(width)
    arr=createLineInside(row, width, height ,arr)
    createHorizontalEdge(width)
    return arr
```

```python
def createRectangule (row, width, height ,arr) :
    
    createHorizontalEdge(width)
    arr=createRectanguleInside(row, width, height ,arr)
    createHorizontalEdge(width)
    return arr
```

```python
def createRectanguleInside(row, width, height,arr):
    x1=int(row[1])
    y1=int(row[2])
    x2=int(row[3])
    y2=int(row[4])
    validations(x1,x2,y1,y2,width,height )
    y1=y1-1
    arr1=arr
    for x in range(y1,y2):
        for y  in range(x1,x2+1):
            arr1[x][y]='X'
            print("x"+str(x+1),"y"+str(y))
    for x in range(y1+1,y2-1):
        for y  in range(x1+1,x2):
            arr1[x][y]=' '
            print("x"+str(x+1),"y"+str(y))
     
    
    
  

    printMatrix(arr)
    return arr
```

```python
def fill(row, width, height ,arr):
    x=int(row[1])
    y=int(row[2])
    for x in range(height):
        for y  in range(width+1):
            if(arr[x][y]==' ') :
                arr[x][y]=row[3]
                
                
    printMatrix(arr)     
    
    return arr
```

```python
def fillBucket (row, width, height ,arr) :
    createHorizontalEdge(width)
    arr=fill(row, width, height ,arr)
    createHorizontalEdge(width)
    return arr
```
